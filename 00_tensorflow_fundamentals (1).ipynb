## Getting Information from Tensors
When dealing with Tensors you want to be aware of the following attributes:

*Shape
*Rank
*Axis of Dim
*Size

```python
#Create a rank 4 tensor (4 dim)
ranked_4_tensor=tf.zeros(shape=(2,3,4,5))
```

```python
ranked_4_tensor
```

```python
ranked_4_tensor.ndim
```

```python
ranked_4_tensor[0]
```

```python
ranked_4_tensor.shape,ranked_4_tensor.ndim,tf.size(ranked_4_tensor)
```

```python
#Get various attributes of our Tensor
print("Datatype of every element:",ranked_4_tensor.dtype)
print("Number of Dimensions (rank):",ranked_4_tensor.ndim)
print("Shape of Tensor:",ranked_4_tensor.shape)
print("Elements along 0 axis:",ranked_4_tensor.shape[0])
print("Elements along Last axis:",ranked_4_tensor.shape[-1])
print("Total no of elements:",tf.size(ranked_4_tensor))
```

###Finding Positional Maximum and Minimum

```python
#Create a new tensor for finding positional minimum and maximum
tf.random.set_seed(42)
F=tf.random.uniform(shape=[50])
F
```

```python

#Find the positional maximum
tf.argmax(F)
```

```python
#Index our largest value
F[tf.argmax(F)]
```

```python

#Max value of F
tf.reduce_max(F)
```

```python

#Check for equality
F[tf.argmax(F)]==tf.reduce_max(F)
```

```python

#Find the positional minimum
tf.argmin(F)
```

```python
#Find the minimum
F[tf.argmin(F)]
```
